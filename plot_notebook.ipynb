```python
from plot import process_files_in_folder, create_histogram
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Process files in the folder
# process_files_in_folder(folder_path)
def plotter(folder_path, n_trials):
    for density in range(n_trials):
    
        # Read the CSV file
        df = pd.read_csv(f"{folder_path}/{str(density)}_results.csv")

        print(df)
    
        # Extract columns for the first set of points
        x1 = df.iloc[:, 4]  # 5th column
        y1 = df.iloc[:, 3]  # 4th column

    
        # Extract columns for the second set of points
        x2 = df.iloc[:, 7]  # 7th column
        y2 = df.iloc[:, 6]  # 6th column
    
        # Plot the first set of points
        plt.scatter(x1, y1, color='blue', label='POMDP')


        print("POMDP")
        for x, y in zip(x1, y1):
            print(f"{x}\t{y}\n")  # Assuming tab-separated values

    
        # Plot the second set of points
        plt.scatter(x2, y2, color='red', label='Policy Iteration')

        print("Policy iteration")
        for x, y in zip(x2, y2):
            print(f"{x}\t{y}\n")  # Assuming tab-separated values
    
        # Add labels and title
        plt.xlabel('X-axis')
        plt.ylabel('Y-axis')
        plt.title('Reward vs communication')
    
        # Add legend
        plt.legend()
    
        # Show the plot
        plt.grid(True)
        plt.show()


        # Extract columns for the first set of points
        y1 = df.iloc[:, 0]  # 4th column
    
        # Extract columns for the second set of points
        y2 = df.iloc[:, 8]  # 6th column

        # Extract columns for the bound set of points
        flag = df.iloc[:,5]
    
        # Plot the first set of points
        plt.plot(y1*1000, color='blue', label='POMDP')
        
        # Plot the first set of points
        plt.plot(y1*1000 + flag, color='green', label='POMDP bound')
    
        # Plot the second set of points
        plt.plot(y2, color='red', label='Policy Iteration')
    
        # Add labels and title
        plt.xlabel('X-axis')
        plt.ylabel('Y-axis')
        plt.title('Reward vs communication')
    
        # Add legend
        plt.legend()
    
        # Show the plot
        plt.grid(True)
        plt.show()
```

```python
# Specify the folder containing the files
folder_path = '10state/estimation/csvs/'
number_of_trials = 5

plotter(folder_path, number_of_trials)
```
